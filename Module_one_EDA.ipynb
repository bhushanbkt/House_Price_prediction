```python
plt.figure(figsize=(15, 6))

# Distribution of exactPrice
plt.subplot(1, 2, 1)
sns.histplot(df['exactPrice'], bins=30, kde=True)
plt.title('Distribution of exactPrice')

# Log-transformed distribution of exactPrice
plt.subplot(1, 2, 2)
sns.histplot(np.log1p(df['exactPrice']), bins=30, kde=True)  # Using log1p to handle zero values
plt.title('Log-transformed Distribution of exactPrice')

plt.show()
```

The distribution of house prices [exactPrice] shows that most houses have lower prices, but there are a few significantly more expensive ones, especially noticeable after applying a log transformation for a clearer view.

```python
# Calculate correlation matrix
correlation_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(20, 16))
sns.heatmap(correlation_matrix[['exactPrice']], annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1)
plt.title('Correlation with exactPrice')
plt.show()
```

```python
# Calculate correlation values
correlation_values = correlation_matrix['exactPrice'].sort_values(ascending=False)

# Plot bar chart
plt.figure(figsize=(12, 8))
correlation_values.drop('exactPrice').plot(kind='bar', color='skyblue')
plt.title('Correlation of Numerical Features with exactPrice')
plt.xlabel('Numerical Features')
plt.ylabel('Correlation Coefficient')
plt.show()
```

```python
# Specify the target column
target_column = "exactPrice"

# Calculate correlations with 'exactPrice' for all columns
correlations = df.corr()[target_column]

# Exclude the 'exactPrice' column itself
correlations = correlations.drop(target_column)

# Print the correlation values
for column, correlation_value in correlations.items():
    print(f"The correlation between '{column}' and '{target_column}' is: {correlation_value:.2f}")
```

```python
#  columns we want to keep
selected_columns = ['exactPrice', "sqftPrice","carpetArea"
    , "bedrooms", "bathrooms",
    "city", "locality", "furnishing",
    "facing", "propertyType", "balconies"
]

# Create a new DataFrame with only the selected columns
new_df = df[selected_columns]

# Display the first few rows of the new DataFrame
new_df.head()
```
